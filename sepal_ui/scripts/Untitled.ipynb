```python
import pandas as pd 
```

```python
path = 'country_code.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df = df.sort_values(by=['country_na'])
df
```

```python
fip_dic = {}
for i, row in df.iterrows():
    name = row['country_na']
    val = row['FIPS 10-4']
    
    fip_dic[name] = val
fip_dic
```
